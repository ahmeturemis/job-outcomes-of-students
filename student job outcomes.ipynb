```python
import pandas as pd
import matplotlib.pyplot as plt #pyplot module used for creating different plots.we need magic to display plots inline
%matplotlib inline              
recent_grads = pd.read_csv("recent-grads.csv") 
recent_grads.iloc[0] #displaying first row formatted as a table





```

```python
raw_data_count = len(recent_grads.index) #calculate index of dataframe and display number of rows
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna() #drop rows containing missing values 
```

```python
cleaned_data_count = len(recent_grads.index) #after cleaning rows contains missing values.we have one less rows than before
print(cleaned_data_count)
```

```python
recent_grads['Sample_size'].hist(bins=25, range=(0,5000)) #use series.hist() to control the binning strategy of histogram
```

```python
cols = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]
fig=plt.figure(figsize=(5,16))
for each in range(1,5):
    ax = fig.add_subplot(4,1,each)
    ax = recent_grads[cols[each]].plot(kind="hist",rot=40)
    
```

```python
cols = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]
fig=plt.figure(figsize=(5,12))
for each in range(4,8):
    ax = fig.add_subplot(4,1,each-3)
    ax = recent_grads[cols[each]].plot(kind="hist",rot=40)
    
```
